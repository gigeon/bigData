```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 검색할 키워드 설정
base_keyword = "20대"
exclude_keywords = "-실형"
final_query = f"{base_keyword} {exclude_keywords}"

# 2. 브라우저 옵션 설정
options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # 브라우저 창을 띄우지 않으려면 주석 해제

# 3. 드라이버 실행
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 4. 네이버 뉴스 검색 URL로 이동
# query={search_keyword} 부분에 우리가 원하는 검색어가 들어갑니다.
url = f"https://search.naver.com/search.naver?where=news&query={final_query}&sort=0"
driver.get(url)

# 5. 로딩 대기 (인터넷 환경에 따라 시간 조절 필요)
time.sleep(3)

# 6. 뉴스 제목 요소들 찾기
news_titles = driver.find_elements(By.CSS_SELECTOR, ".sds-comps-text-type-headline1")

print(news_titles)

# 7. 반복문으로 제목과 링크 추출
for news in news_titles:
    title = news.text                # 기사 제목 (텍스트)

    print(f"제목: {title}")
    print("-" * 30)

# 8. 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 똑똑한 키워드 설정 (핵심!)
# '20대'를 검색하되 '주식', '특징주', '포토'는 뺍니다.
base_keyword = "20대 국민연금"

# 2. 브라우저 설정
options = webdriver.ChromeOptions()
# options.add_argument("--headless")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 3. URL 이동 (sort=0은 관련도순, sort=1은 최신순)
# 관련도순으로 봐야 낚시성 기사가 좀 더 걸러집니다.
url = f"https://search.naver.com/search.naver?where=news&query={base_keyword}&sort=1"
driver.get(url)

time.sleep(2)

# 4. 결과 출력
news_titles = driver.find_elements(By.CSS_SELECTOR, ".news_tit")

print(f"--- '{final_query}' 크롤링 결과 ---")

for news in news_titles:
    title = news.text
    link = news.get_attribute("href")

    # (선택사항) 파이썬 내부에서 한 번 더 필터링 할 수도 있습니다.
    if "속보" in title: # 제목에 '속보'가 들어가면 건너뛰기 예시
        continue

    print(f"제목: {title}")
    print(f"링크: {link}")
    print("-" * 30)

driver.quit()
```
