```python
import pandas as pd

# CSV 로드
df = pd.read_csv("dir/korea_person.csv")

# 컬럼명 통일
df.columns = ["age_group", "year", "population"]

wide = (
    df.pivot(index="year", columns="age_group", values="population")
    .sort_index()
)

age_flow = {
    "5~9": "10~14",
    "10~14": "15~19",
    "15~19": "20~24",
    "20~24": "25~29",
    "25~29": "30~34",
    "30~34": "35~39",
    "35~39": "40~44",
    "40~44": "45~49",
    "45~49": "50~54",
    "50~54": "55~59",
    "55~59": "60+",
    "60+": "60+"
}

wide
```

```python
survival_rates = {}

for src, dst in age_flow.items():
    rates = []

    for year in wide.index:
        if year + 5 in wide.index:
            src_pop = wide.loc[year, src]
            dst_pop = wide.loc[year + 5, dst]

            if src_pop > 0:
                rates.append(dst_pop / src_pop)

    survival_rates[src] = sum(rates) / len(rates)

survival_rates
```

```python
forecast_df = wide.copy()

for year in range(2025, 2031):
    forecast_df.loc[year] = forecast_df.loc[year - 1]

    for src, dst in age_flow.items():
        forecast_df.loc[year, dst] = int(
            forecast_df.loc[year - 1, src] * survival_rates[src]
        )
```
